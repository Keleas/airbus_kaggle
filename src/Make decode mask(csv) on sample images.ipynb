## Создать csv формат по одним маскам

```python
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from skimage.data import imread
from skimage.morphology import label
```

```python
## вспомогательные функции для encode/decode 

def multi_rle_encode(img, **kwargs):
    '''
    Encode connected regions as separated masks
    '''
    labels = label(img[0,:,:,:])
    if img.ndim > 2:
        return [rle_encode(np.sum(labels==k, axis=2), **kwargs) for k in np.unique(labels[labels>0])]
    else:
        return [rle_encode(labels==k, **kwargs) for k in np.unique(labels[labels>0])]

# ref: https://www.kaggle.com/paulorzp/run-length-encode-and-decode
def rle_encode(img, min_max_threshold=1e-3, max_mean_threshold=None):
    '''
    img: numpy array, 1 - mask, 0 - background
    Returns run length as string formated
    '''
    if np.max(img) < min_max_threshold:
        return '' ## no need to encode if it's all zeros
    if max_mean_threshold and np.mean(img) > max_mean_threshold:
        return '' ## ignore overfilled mask
    pixels = img.T.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)
```

```python
test_img_dir = 'F:\\Downloads\\all\\mono(true)_finetune\\'
images = os.listdir(test_img_dir)
```

```python
pred_rows = []
for i in range(len(images)):
    test_img = imread(test_img_dir + '\\' + images[i])
    test_img = test_img.reshape(1,768,768,1)/255.0
    rles = multi_rle_encode(test_img)
    if len(rles) > 0:
        for rle in rles:
            # перегоняем все фото под JPG для csv
            pred_rows += [{'ImageId': images[i][:len(images[i])-4] + '.jpg',
                           'EncodedPixels': rle}]
    else:
        pred_rows += [{'ImageId': images[i][:len(images[i])-4] + '.jpg', 'EncodedPixels': None}]
```

```python
sub = pd.DataFrame(pred_rows, columns=['ImageId', 'EncodedPixels'])
sub.to_csv('F:\\Downloads\\all\\Submissions\\\mono(true)_finetune.csv', index=False)
```
